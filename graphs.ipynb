```python
def create_graph(data, columns):
    filtered = []
    for x in xrange(len(columns)):
        filtered.append([])
    for row in data:
        for i, column in enumerate(columns):
            if not column[0] in row:
                continue
            start = row[column[0]]
            for req in column[1:]:
                start = start[req]
            filtered[i].append(start)
    for x in xrange(1, len(columns)):
        #pylab.subplot(1, len(columns)-1, x)
        pylab.plot(filtered[0], filtered[x], label=str(columns[x][-1]))
```

```python
from simplejson import load
data = load(open("going_up.json"))
```

```python
create_graph(data, [['time'], ['rawMeasures', 'accelerometers', 'x'], ['rawMeasures', 'accelerometers', 'y'], ['rawMeasures', 'accelerometers', 'z']])
legend()
```
